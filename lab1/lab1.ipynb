6. Написати програму побудови лінійного графіку. Значення х задати у
вигляді списку. Список значень у отримайте використовуючи функцію, яка
збільшить значення х на 2 (або можна вибрати ваш варіант). Задати
параметри діаграми: назву, підписи осей, легенду, сітку. Стиль графіку,
колір та розмір текстів задайте самостійно.

```python
x_values = [1, 12, 5, 9]

def calculate_y(x):
    return x + 2
y_values = [calculate_y(x) for x in x_values]
plt.figure(figsize=(6, 4))
plt.plot(x_values, y_values, label='y = x + 2', color='blue', linestyle='--', marker='o')
plt.title('Лінійний графік функції y = x + 2', fontsize=12, color='darkred')
plt.xlabel('Значення X', fontsize=12, color='purple')
plt.ylabel('Значення Y', fontsize=12, color='purple')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

7. Написати програму побудови графіків функцій розміщуючи їх на різних
полях однієї фігури (subplot()). Стилі графіків вибрати на свій розсуд.

```python
x1 = np.linspace(-2, 5, 400)
y1 = x1 * np.sin(5 * x1)

# Друга функція: Y(x) = 5 * sin(1/x) * (cos(x^2)^3), x = [-4, 4] (уникаємо x = 0 через поділ на нуль)
x2 = np.linspace(-4, -0.1, 200).tolist() + np.linspace(0.1, 4, 200).tolist()  # Виключаємо нуль
x2 = np.array(x2)
y2 = 5 * np.sin(1 / x2) * (np.cos(x2**2))**3

plt.figure(figsize=(10, 6))

# Перший графік на першому підграфіку
plt.subplot(1, 2, 1)
plt.plot(x1, y1, label=r'$y = x \cdot \sin(5 \cdot x)$', color='blue', linestyle='--')
plt.title('Графік 1: y = x * sin(5 * x)', fontsize=12, color='darkblue')
plt.xlabel('x', fontsize=10)
plt.ylabel('y', fontsize=10)
plt.grid(True)
plt.legend()

# Другий графік на другому підграфіку
plt.subplot(1, 2, 2)
plt.plot(x2, y2, label=r'$y = 5 \cdot \sin(1/x) \cdot (\cos(x^2))^3$', color='green', linestyle='-')
plt.title('Графік 2: y = 5 * sin(1/x) * (cos(x²))³', fontsize=12, color='darkgreen')
plt.xlabel('x', fontsize=10)
plt.ylabel('y', fontsize=10)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

9. Написати програму побудови стовпчикової гістограми частоти появи
голосних літер у тексті (текст зчитується із текстового файлу, розмір
тексту до 100 слів). Отриману гістограму зберегти у png файл.

```python
from collections import Counter

vowels = "AEIOUYaeiouy"

def count_vowels(text):
    return Counter([char for char in text if char in vowels])

with open("./input_text.txt", "r", encoding="utf-8") as file:
    text = file.read()

vowel_count = count_vowels(text)

letters = list(vowel_count.keys())
counts = list(vowel_count.values())

plt.figure(figsize=(8, 4))
plt.bar(letters, counts, color='skyblue')

plt.title('Frequency of appearance of vowels in the text', fontsize=12)
plt.xlabel('Vowel letters', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.savefig('./task9.png')

plt.show()
```

10. За даними таблиці написати програму побудови кругової діаграми. Для
даної діаграми відобразити назву (взяти з заголовку таблиці), значення,
легенду. Задати колір для фірм на свій вибір. Одну із частин розмістити

Валовий прибуток підприємств, тис. грн.

| A    | B     | C     |
| ---  | ---   | ---   |
| 2310 | 3145  | 1520  |

```python
labels = ['A', 'B', 'C']
sizes = [2310, 3145, 1520]
colors = ['skyblue', 'lightgreen', 'lightcoral']

explode = (0, 0.1, 0)

plt.figure(figsize=(5, 5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
        autopct='%1.1f%%', startangle=90)

plt.title('Валовий прибуток підприємств, тис. грн.', fontsize=12)

plt.legend(labels, title="Підприємства", loc="upper right")

plt.savefig('./task10.png')

plt.show()
```

11. Використовуючи програмний код з попереднього завдання, отримати кругову діаграму у вигляді бублика. Результат зберегти у файл.

```python
labels = ['A', 'B', 'C']
sizes = [2310, 3145, 1520]
colors = ['skyblue', 'lightgreen', 'lightcoral']

explode = (0, 0.1, 0)

plt.figure(figsize=(5, 5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
        autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.1})

plt.title('Валовий прибуток підприємств, тис. грн.', fontsize=12)

plt.legend(labels, title="Підприємства", loc="upper right")

plt.savefig('./task11.png')

plt.show()
```

12. Написати програму побудови групової гістограми за даними таблиці. Всі параметри гістограми мають бути задані. Результат зберегти у файл.

| Об’єми продажу підприємств, тис. грн. | 1 кв. | 2 кв. | 3 кв. | 4 кв. |
|---------------------------------------|-------|-------|-------|-------|
| A                                     | 470   | 684   | 350   | 806   |
| B                                     | 787   | 586   | 980   | 792   |
| C                                     | 320   | 440   | 360   | 400   |


```python
quarters = ['1 кв.', '2 кв.', '3 кв.', '4 кв.']
sales_A = [470, 684, 350, 806]
sales_B = [787, 586, 980, 792]
sales_C = [320, 440, 360, 400]

bar_width = 0.25
index = np.arange(len(quarters))


plt.figure(figsize=(10, 4))
bar1 = plt.bar(index - bar_width, sales_A, bar_width, label='A', color='skyblue')
bar2 = plt.bar(index, sales_B, bar_width, label='B', color='lightgreen')
bar3 = plt.bar(index + bar_width, sales_C, bar_width, label='C', color='lightcoral')

plt.xlabel('Квартал', fontsize=12)
plt.ylabel('Обсяг продажу (тис. грн.)', fontsize=12)
plt.title('Групова гістограма обсягів продажу підприємств', fontsize=12)
plt.xticks(index, quarters)
plt.legend(title='Підприємства')

plt.savefig('./task12.png')

plt.show()
```

13. Використовуючи програмний код попереднього завдання отримати горизонтальну групову діаграму та складену групову діаграму. Результати зберегти у файли.

```python
quarters = ['1 кв.', '2 кв.', '3 кв.', '4 кв.']
sales_A = [470, 684, 350, 806]
sales_B = [787, 586, 980, 792]
sales_C = [320, 440, 360, 400]

bar_width = 0.25
index = np.arange(len(quarters))


plt.figure(figsize=(10, 4))
bar1 = plt.barh(index - bar_width, sales_A, bar_width, label='A', color='skyblue')
bar2 = plt.barh(index, sales_B, bar_width, label='B', color='lightgreen')
bar3 = plt.barh(index + bar_width, sales_C, bar_width, label='C', color='lightcoral')

plt.xlabel('Квартал', fontsize=12)
plt.ylabel('Обсяг продажу (тис. грн.)', fontsize=12)
plt.title('Групова гістограма обсягів продажу підприємств', fontsize=12)
plt.xticks(index, quarters)
plt.legend(title='Підприємства')

plt.savefig('./task13p1.png')

plt.show()
```

```python
quarters = ['1 кв.', '2 кв.', '3 кв.', '4 кв.']
sales_A = [470, 684, 350, 806]
sales_B = [787, 586, 980, 792]
sales_C = [320, 440, 360, 400]

bar_width = 0.5
index = np.arange(len(quarters))

plt.figure(figsize=(10, 4))

plt.bar(index, sales_A, bar_width, label='A', color='skyblue')
plt.bar(index, sales_B, bar_width, bottom=sales_A, label='B', color='lightgreen')
plt.bar(index, sales_C, bar_width, bottom=np.array(sales_A) + np.array(sales_B), label='C', color='lightcoral')

plt.xlabel('Квартал', fontsize=12)
plt.ylabel('Обсяг продажу (тис. грн.)', fontsize=12)
plt.title('Складена гістограма обсягів продажу підприємств', fontsize=12)
plt.xticks(index, quarters)
plt.legend(title='Підприємства')

plt.savefig('./task13p2.png')

plt.show()
```
